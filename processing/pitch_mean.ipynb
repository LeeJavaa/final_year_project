```python
import librosa
from pydub import AudioSegment
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("../data/dataset/validation/validation.csv")
pitch_mean_df = pd.DataFrame(columns=['id', 'pitch_mean'])
```

```python
def convert_mp4_to_wav(input_file, output_file):
    audio = AudioSegment.from_file(input_file, format="mp4")
    audio.export(output_file, format="wav")
```

```python
for audio_id in df['id']:
    # Import audio file
    audio_file = f'../data/audio/validation/{audio_id}.mp4'
    convert_mp4_to_wav(audio_file, "./audio.wav")
    y, sr = librosa.load('./audio.wav')

    # Calculate pitch mean
    frequencies = librosa.yin(y=y, sr=sr, fmin=60, fmax=200)
    mean_freq = np.mean(frequencies)
    print(mean_freq)

    # Add row to df
    new_row = {'id': audio_id, 'pitch_mean': mean_freq}
    pitch_mean_df = pd.concat([pitch_mean_df, pd.DataFrame([new_row])], ignore_index=True)

    # print out df
    print(pitch_mean_df)
```

```python
pitch_mean_df.to_csv("../data/dataset/validation/validation_audio2.csv", header=True, index=False)
```
